### Problem 8.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
### (a)
college = pd.read_csv('data/College.csv', sep=',')
```

```python
### (b)
college = college.rename({'Unnamed: 0': 'College'}, axis=1)
college = college.set_index('College')
college
```

```python
### (c)
college.describe()
```

```python
### (d)
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']]);
```

```python
### (e)
fig, ax = plt.subplots(figsize=(8, 8))
college.boxplot('Outstate', by='Private', ax = ax);
```

```python
### (f)
college['Elite'] = pd.cut(college['Top10perc'],
                          [0, 50, 100],
                          labels=['No', 'Yes'])
college['Elite'].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
college.boxplot('Outstate', by='Elite', ax=ax);
```

```python
### (g)
fig, axes = plt.subplots(2, 2)
college.hist('Top10perc', bins=10, ax=axes[0][0]);
college.hist('Outstate', bins=12, ax=axes[0][1]);
college.hist('Room.Board', bins=14, ax=axes[1][0]);
college.hist('Grad.Rate', bins=16, ax=axes[1][1]);
```
